**AIM** : Implementation of Prediction Algorithm (Linear Regression)

**INTRODUCTION** : This experiment implements linear regression from scratch using gradient descent optimization. Linear regression models the relationship between variables using the equation y = wx + b, where model parameters (w, b) are optimized to minimize prediction error without relying on pre-built machine learning libraries.

**PROCEDURE:**

1. Setup: Created a LinearRegression class with configurable learning rate and iterations; generated synthetic test data

2. Core Implementation: Developed prediction function, mean squared error calculation, and gradient descent optimization

3. Training Process: Initialized parameters at zero, iteratively updated them based on error gradients, and tracked convergence

4. Evaluation: Calculated RÂ² score and visualized both the regression line and cost history

```python
import numpy as np
import matplotlib.pyplot as plt

class LinearRegression:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None
        self.cost_history = []

    def fit(self, X, y):
        """
        Train the linear regression model using gradient descent

        Parameters:
        X (array): Training data of shape (n_samples, n_features)
        y (array): Target values of shape (n_samples,)
        """
        # Initialize parameters
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient descent
        for i in range(self.n_iterations):
            # Calculate predictions
            y_predicted = self._predict(X)

            # Calculate gradients
            dw = (1/n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1/n_samples) * np.sum(y_predicted - y)

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # Calculate cost
            cost = self._calculate_cost(y, y_predicted)
            self.cost_history.append(cost)

            # Print cost every 100 iterations
            if (i+1) % 100 == 0:
                print(f'Iteration {i+1}: Cost = {cost}')

    def _predict(self, X):
        """
        Make predictions using the current model parameters
        """
        return np.dot(X, self.weights) + self.bias

    def predict(self, X):
        """
        Make predictions for new data
        """
        return self._predict(X)

    def _calculate_cost(self, y_true, y_pred):
        """
        Calculate the mean squared error cost
        """
        n_samples = len(y_true)
        cost = (1/n_samples) * np.sum((y_pred - y_true)**2)
        return cost

    def r_squared(self, X, y):
        """
        Calculate the coefficient of determination (R^2)
        """
        y_pred = self.predict(X)
        ss_total = np.sum((y - np.mean(y))**2)
        ss_residual = np.sum((y - y_pred)**2)
        r2 = 1 - (ss_residual / ss_total)
        return r2

    def plot_cost_history(self):
        """
        Plot the cost function over iterations
        """
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, len(self.cost_history) + 1), self.cost_history)
        plt.xlabel('Iterations')
        plt.ylabel('Cost')
        plt.title('Cost Function Over Iterations')
        plt.grid(True)
        plt.show()


# Example usage
if __name__ == "__main__":
    # Generate some sample data
    np.random.seed(42)
    X = 2 * np.random.rand(100, 1)
    y = 4 + 3 * X + np.random.randn(100, 1)
    # Reshape y to (100,) to match the expected shape
    y = y.flatten()

    # Create and train the model
    model = LinearRegression(learning_rate=0.01, n_iterations=1000)
    model.fit(X, y)

    # Make predictions
    X_test = np.array([[0], [2]])
    y_pred = model.predict(X_test)

    # Calculate R^2
    r2 = model.r_squared(X, y)
    print(f"R^2 Score: {r2}")

    # Plot results
    plt.figure(figsize=(10, 6))
    plt.scatter(X, y, color='blue', label='Data points')
    plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title('Linear Regression Example')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Plot cost history
    model.plot_cost_history()
```

**CONCLUSION** :

The implementation successfully approximated the underlying data relationship through gradient descent optimization. This approach provided computational efficiency while maintaining full visibility into the algorithm's workings. The implementation demonstrates how linear regression fundamentals work while serving as both a practical tool and educational resource. The approach could be extended with regularization, feature normalization, and advanced optimization techniques for more complex applications.
